# Create function Scraping URL

````python
def get_url_content(i_page):
    url = f'https://www.moj.go.th/ministertalk/contacts/index?Contacts_page={i_page}&ajax=yw0'
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")
    div = soup.find('div' , {'class':'items'})
    
    
    links = []
    ios_categories_links=[]
    hashmap_genre_link ={}
        
    href = div.find_all('a', href=True)
    for j in href:
    # print(j['href'])
        links.append(j['href'])
        
    df_url = pd.DataFrame(links)
    df_url.columns = ["URL_name"]
    
    return df_url
````

````python
page = np.arange(1, 7).tolist()
df_all = pd.DataFrame()

for i_page in page:
    try:
        df_url = get_url_content(i_page)
        df_all = pd.concat([df_all, df_url], axis=0, sort=False)
        print(f'Done getting data from page {i_page}')
    except:
        print(f'Cannot get data from page {i_page}')

df_all.to_excel('moj_url.xlsx', index=False)
````

## Loop Scraping text

````python
moj_url = pd.read_excel("moj_url_ver2.xlsx", header=[0])
````

````python
moj_url 
````

````python
def get_QA(i):
    URL = moj_url['URL_name'][i]
    r = requests.get(URL)
    soup = BeautifulSoup(r.text, "html.parser")
    div = soup.find_all('div' , {'class':'entry-content'})

    Question = div[1].text
    Anw = div[2].text
                
    df = pd.DataFrame([[Question,Anw]],columns=['คำถาม','คำตอบ'])
    return(df)
````

````python
df_QA = pd.DataFrame()

for i in range(len(moj_url)):
    try:
        df = get_QA(i)
        df_QA = pd.concat([df, df_QA], axis=0, sort=False)
        print(f'Done getting data from page {i}')
    except:
        print(f'Cannot get data from page {i}')   
````

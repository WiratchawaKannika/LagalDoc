```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
def get_table_content(i_page):
    url = f'http://thanaiphorn.com/boards/index.php?board=1.{i_page}0'
    source = requests.get(url)
    
    soup = BeautifulSoup(source.text, 'lxml', from_encoding='utf-8')
    text_body = soup.find('tbody')
    span_tags = text_body.find_all('span')
    rows=[]
    for a in span_tags: 
        a_tags=a.find('a')
        link=a_tags['href']
        rows.append((link))
    df = pd.DataFrame(rows, columns=['Links'])
    return(df)
```

```python
n = 0
page = []
while n <= 52 :
    page.append(n)
    n = n + 2
print(page)
```

```python
n = 0
page = []
while n <= 52 :
    page.append(n)
    n = n + 2

df_all = pd.DataFrame()
for i_page in page:
    try:
        df = get_table_content(i_page)
        df_all = pd.concat([df_all, df], axis=0, sort=False)
        print(f'Done getting data from page {i_page}')
    except:
        print(f'Cannot get data from page {i_page}')

df_all.to_excel('Thanaiphorn_Links.xlsx', index=False)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

#สร้างฟังก์ชันสำหรับ scrape URL
def get_table_content(i_page):
    url = f'http://thanaiphorn.com/boards/index.php?board=1.{i_page}0'
    source = requests.get(url)
    
    soup = BeautifulSoup(source.text, 'lxml', from_encoding='utf-8')
    text_body = soup.find('tbody')
    span_tags = text_body.find_all('span')
    rows=[]
    for a in span_tags: 
        a_tags=a.find('a')
        link=a_tags['href']
        rows.append((link))
    df = pd.DataFrame(rows, columns=['Links'])
    return(df)
#for loop เพื่อ scrape URL จากทุกหน้าเว็บ
n = 0
page = []
while n <= 52 :
    page.append(n)
    n = n + 2

df_all = pd.DataFrame()
for i_page in page:
    try:
        df = get_table_content(i_page)
        df_all = pd.concat([df_all, df], axis=0, sort=False)
        print(f'Done getting data from page {i_page}')
    except:
        print(f'Cannot get data from page {i_page}')
#เก็บ URL เป็นไฟล์ Excel
df_all.to_excel('Thanaiphorn_Links.xlsx', index=False)

#อ่านไฟล์ Excel URL
url_file = pd.read_excel (r'Thanaiphorn_Links.xlsx')
#สร้างฟังก์ชันสำหรับ scrape Q&A
def get_table_QandA(i):
    URL = url_file['Links'][i]
    S = requests.get(URL)

    Soup = BeautifulSoup(S.text, 'lxml', from_encoding='utf-8')
    Form = Soup.find('form' , {'id':'quickModForm'})
    Div = Form.find_all('div' , {'class':'inner'})
    Q = Div[0].text

    Member = Form.find_all('li' , {'class':'postgroup'})
    for j in range(len(Div)):
        if j > 0 :
            if Member[j].text == 'Hero Member':
                A = Div[j].text
                break
    df = pd.DataFrame([[Q,A]],columns=['คำถาม','คำตอบ'])
    return(df)
#for loop เพื่อ scrape Q&A จากทุกหน้าเว็บ
QandA_df = pd.DataFrame()
for i in range(len(url_file)):
    try:
        df = get_table_QandA(i)
        QandA_df = pd.concat([QandA_df, df], axis=0, sort=False)
        print(f'Done getting data from page {i}')
    except:
        print(f'Cannot get data from page {i}')
QandA_df.reset_index(drop=True,inplace=True)
#เก็บ Q&A เป็นไฟล์ Excel
QandA_df.to_excel('Thanaiphorn_Q&A.xlsx', index=False)
```

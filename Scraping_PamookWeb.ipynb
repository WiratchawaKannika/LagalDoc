```python
def get_url_content(i_page):
    url = f'https://www.pamook.com/index.php?lay=show&ac=webboard&WBntype=1&Category=pamookcom&sort=123&sorttype=&thispage={i_page}'
    
    links = []
    ios_categories_links=[]
    hashmap_genre_link ={}

    response = requests.get(url)
    soup = BeautifulSoup(response.text,"html.parser")
    body = soup.find(class_='content')
    tables = body.find('table', class_='webboard-all-res')
    href = tables.find_all('a', href=True)
    
    for j in href:
    # print(j['href'])
        links.append(j['href'])
        
    pamook_url = pd.DataFrame(links)
    pamook_url.columns = ["URL_name"]
    
    return pamook_url
```

## Loop Scraping text 

```python
import pandas as pd
pamook_url = pd.read_excel("pamook_url_real.xlsx", header=[0])
```

```python
pamook_url
```

```python
def get_QA(i):
    URL = pamook_url['URL_name'][i]
    r_test = requests.get(URL)
    encoding_test = r_test.encoding if 'charset' in r_test.headers.get('span', '').lower() else None
    elements = BeautifulSoup(r_test.content, from_encoding=encoding_test)
    table = elements.find_all('table', class_ ='webboard_display_small tplImport-table')
    td = elements.find_all('td', class_ ='content')

    Question = td[1].text
    Anw = td[5].text
                
    df_data = pd.DataFrame([[Question,Anw]],columns=['คำถาม','คำตอบ'])
    return(df_data)
```

```python
get_QA(3)
```

```python
range(len(pamook_url))
```

```python
df_QA = pd.DataFrame()

for i in range(len(pamook_url)):
    try:
        df_data = get_QA(i)
        df_QA = pd.concat([df_data, df_QA], axis=0, sort=False)
        print(f'Done getting data from page {i}')
    except:
        print(f'Cannot get data from page {i}')   
```

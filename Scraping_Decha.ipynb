```python
def get_table_QandA(link_i):
    URL = f'https://www.decha.com/board/group/BoardLaw/{link_i}'
    S = requests.get(URL)
    Soup = BeautifulSoup(S.text, 'lxml', from_encoding='utf-8')
    
    Q = Soup.find('div' , {'class':'detail_board_flukydu'}).text
   
    for j in range(len(Soup.find_all('div' , {'class':'postmessage-box cmt-h'}))):
        if Soup.find_all('div' , {'class':'postmessage-box cmt-h'})[j].find('h6').text.split()[1]+Soup.find_all('div' , {'class':'postmessage-box cmt-h'})[j].find('h6').text.split()[2] == 'อ.เดชากิตติวิทยานันท์':
            y1 = j
            A = ''
            for i in range(len(Soup.find_all('div' , {'class':'postmessage-box cmt-h'})[y1].find_all('p'))) :
                x = Soup.find_all('div' , {'class':'postmessage-box cmt-h'})[y1].find_all('p')[i].text
                A = A+x
            break
        elif Soup.find_all('div' , {'class':'postmessage-box cmt-h'})[j].find('h6').text.split()[1]+Soup.find_all('div' , {'class':'postmessage-box cmt-h'})[j].find('h6').text.split()[2] == 'ทีมงานทนายคลายทุกข์(สมาชิก)':
            y2 = j
            A = ''
            for i in range(len(Soup.find_all('div' , {'class':'postmessage-box cmt-h'})[y2].find_all('p'))) :
                x = Soup.find_all('div' , {'class':'postmessage-box cmt-h'})[y2].find_all('p')[i].text
                A = A+x
            break 
        
        elif Soup.find_all('div' , {'class':'postmessage-box cmt-h'})[j].find('h6').text.split()[1]+Soup.find_all('div' , {'class':'postmessage-box cmt-h'})[j].find('h6').text.split()[2] == 'มโนธรรมเจษฎาสาธุชน':
            y3 = j
            A = ''
            for i in range(len(Soup.find_all('div' , {'class':'postmessage-box cmt-h'})[y3].find_all('font'))) :
                x = Soup.find_all('div' , {'class':'postmessage-box cmt-h'})[y3].find_all('font')[i].text
                A = A+x
            break
    df = pd.DataFrame([[Q,A]],columns=['คำถาม','คำตอบ'])
    return(df)
```

```python
QandA_df = pd.DataFrame()
for i in range(23002):
    try:
        df = get_table_QandA(i)
        QandA_df = pd.concat([QandA_df, df], axis=0, sort=False)
        print(f'Done getting data from page {i}')
    except:
        print(f'Cannot get data from page {i}')   
```

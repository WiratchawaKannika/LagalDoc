````python
def get_table_pamook(url):
    source = requests.get(url)
    encoding_test = source.encoding if 'charset' in source.headers.get('span', '').lower() else None
    soup_test = BeautifulSoup(source.content, from_encoding=encoding_test )
    body = soup_test.find(class_='webboard_frame_small tplImport-table')
    elements = body.find_all('table', class_='webboard_display_small tplImport-table')

    ls = [] # list to store text of all elements in all the rows.
    # list to store text of all the elements in this row:
    ls_elements_in_row = []
    
    for element in elements:
        text = element.text
        ls_elements_in_row += [text]
    
    ls += [ls_elements_in_row]

    df = pd.DataFrame(ls)

    return df
````

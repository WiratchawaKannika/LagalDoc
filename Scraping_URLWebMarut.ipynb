```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

- สร้าง function เพื่อเอา URL ของแต่ละหน้า จากเว็บ

```python
def get_url_content(i_page):
    url = f'http://www.marutbunnag.com/webboard/index.php?language=thai&pageno={i_page}'
    response = requests.get(url)
    html_soup = BeautifulSoup(response.text,"html.parser")
    div = html_soup.find('div', id='contentdiv')
    categories_class = div.find_all(class_='topiclist')
    table = div.find('table')
    
    links = []
    ios_categories_links=[]
    hashmap_genre_link ={}
    
    href = table.find_all('a', href=True)
    for j in href:
    # print(j['href'])
        links.append(j['href'])
        
    df_url = pd.DataFrame(links)
    df_url.columns = ["URL_name"]
    
    return df_url
```

```python
get_url_content('http://www.marutbunnag.com/webboard/index.php?language=thai&pageno=3&pagesize=40')
```

- LOOP เพื่อเก็บ URL จากทุกหน้า Web

```python
df_all = pd.DataFrame()
for i_page in range(0, 14):
    try:
        df_url = get_url_content(i_page)
        df_all = pd.concat([df_all, df_url], axis=0, sort=False)
        print(f'Done getting data from page {i_page}')
    except:
        print(f'Cannot get data from page {i_page}')

df_all.to_excel('table_url.xlsx', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_table_content(url):
    source = requests.get(url)
    soup = BeautifulSoup(source.content, 'lxml', from_encoding='utf-8')
    div = soup.find('div', id='contentdiv')
    elements = div.find_all(class_='mkbb')

    ls = [] # list to store text of all elements in all the rows.
    # list to store text of all the elements in this row:
    ls_elements_in_row = []
    
    for element in elements:
        text = element.text
        ls_elements_in_row += [text]
    
    ls += [ls_elements_in_row]

    df = pd.DataFrame(ls)

    return df
```

- LOOP เพื่อเก็บ Text จากทุกหน้า ส่งออกในรูปตาราง ชื่อ Scraping_marut.xlsx

```python
for i in range(len(table_url["URL_name"])):
    url = table_url["URL_name"][i]
    try:
        df     = get_table_content(url)
        table_all  = pd.concat([table_all, df], axis=0, sort=False,\
                               ignore_index=True, join="outer")
        print(f"Done scraping {i}")
        table_all.to_pickle("temp.pkl")
           
    except Exception as e:
        print(f"Cannot scrape {i}")
        print(str(e))

table_all.to_excel('Scraping_marut.xlsx', index=False)
```

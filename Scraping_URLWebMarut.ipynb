- สร้าง function เพื่อเอา URL ของแต่ละหน้า จากเว็บ

```python
def get_url_content(i_page):
    url = f'http://www.marutbunnag.com/webboard/index.php?language=thai&pageno={i_page}&pagesize=40'
    response = requests.get(url)
    html_soup = BeautifulSoup(response.text,"html.parser")
    div = html_soup.find('div', id='contentdiv')
    categories_class = div.find_all(class_='topiclist')
    table = div.find('table')
    
    links = []
    ios_categories_links=[]
    hashmap_genre_link ={}
    
    href = table.find_all('a', href=True)
    for j in href:
    # print(j['href'])
        links.append(j['href'])
        
    df_url = pd.DataFrame(links)
    df_url.columns = ["URL_name"]
    
    return df_url
```

```python
get_url_content('http://www.marutbunnag.com/webboard/index.php?language=thai&pageno=3&pagesize=40')
```

- LOOP เพื่อเก็บ URL จากทุกหน้า Web

```python
import numpy as np
page = np.arange(1, 14).tolist()
```

```python
page = np.arange(1, 14).tolist()

df_all = pd.DataFrame()
for i_page in page:
    try:
        df_url = get_url_content(i_page)
        df_all = pd.concat([df_all, df_url], axis=0, sort=False)
        print(f'Done getting data from page {i_page}')
    except:
        print(f'Cannot get data from page {i_page}')

df_all.to_excel('Marut_url.xlsx', index=False)
```

# Loop Scraping

```python
def get_table_content(i):
    URL = table_url['URL_name'][i]
    r = requests.get(URL)
    r.encoding = 'utf-8' # กำหนด encoding กันปัญหาการถอดรหัสผิดแบบ
    soup = BeautifulSoup(r.text,'lxml')
    Ques = soup.find_all('div' , {'class':'mkbb'})
    
    Q = Ques[0].text
    A = Ques[1].text
   
    df = pd.DataFrame([[Q,A]],columns=['คำถาม','คำตอบ'])
    return(df)
```
